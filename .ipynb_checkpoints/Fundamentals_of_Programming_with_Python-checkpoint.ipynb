## Variables and Data Types

```python
name = "Karan"
age = 25
height = 5.8
is_student = True
print(type(name))
print(type(age))
print(type(height))
print(type(is_student))
```

## Operators in Python

```python
a = 10
b = 3
print(a + b)
print(a - b)
print(a * b)
print(a / b)
print(a % b)
```

```python
print(a == b)
print(a != b)
print(a > b)
print(a < b)
```

```python
x = True
y = False
print(x and y)
print(x or y)
print(not x)
```

## Conditional Statements (if, elif, else)

```python
marks = 75
if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
else:
    print("Grade C")
```

## Loops in Python

```python
for i in range(5):
    print("Number:", i)
```

```python
count = 0
while count < 5:
    print("Count:", count)
    count += 1
```
